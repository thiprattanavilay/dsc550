Data visualization is important people, this helps tell a clear story of their data when presented. When visualizing the number of values, a certain category has, a Pie charts is the best visualization since it easily shows number of values that each category contains. Pie charts are the most common method for visualizing this sort of data since each value can be compared with other values. The second would be the bar charts because it easily shows number of values that each category contains just like Pie chart. Lastly it would be the line chart.

Bar charts are often confused with histograms and looking at the bar charts above can help you understand why. In the example above, we have one qualitative variable, city, and a quantitative variable, which are the total toys. Each city has a corresponding value on the horizontal axis, which represents the total toys. As easy to ready, it can be large as you scale which makes it harder to read.

Line charts are the worst at displaying the count of categorical data. Line charts imply a continuation of the data, like a time series. Since this type of data has no ordinal values, we should not use a line to demonstrate the count. Line charts are also not designed for displaying exact values since the y axis doesn’t typically start at 0, unlike bar charts. Line charts typically show relative changes within the series, so using like charts to display this sort of data is a very poor decision.

Pie charts are by far superior when every portion sums up to exactly 100%, and you could compare every portion easily if you sort them. You could also compare multiple pie charts side by side if percentages of a recurring category matter. It also doesn’t need a scale indicator, everything is 100% in full circle. It makes sense with all the data points are displayed accurately.

## 1. Create an array of size [10,50] and use the random function to fill it with integers between 1 and 500 (inclusive). Load it to a DataFrame.

#### importing libraries

```python
import numpy as np # np for short hand name for numpy
import pandas as pd # pd for short hand name for pandas
```

```python
## Using numpy library to generate 500 random numbers into an array of size 10,50
nums = np.random.randint(1,501, size=(10, 50))
```

**Load it to a DataFrame.**

```python
## Loading the array into a dataframe using pandas library. Column names will be 1-50
df = pd.DataFrame(nums, columns=np.arange(1,51))

## Ensuring index will start at 1
df.index = np.arange(1, len(df) + 1)
df
```

## 2. Calculate the sum of each row; the sum of each column; and the sum of all entries.


```python
## loop that sums all 10 rows.

for i,j in enumerate(df.sum(axis=1), start = 1):
    print("Row {:>2} Sum: {}".format(i, j))
```

**b the sum of each column**

```python
## loop that sums all 10 columns.

for i,j in enumerate(df.sum(axis=0), start = 1):
    print("Column {:>2} Sum: {}".format(i,j))
```

**c and the sum of all entries**

```python
## Using df.values.sum() to sum all the values in the dataframe

print("Total Sum: {}".format(df.values.sum()))
```

## 3. Determine the minimum; maximum; average; and standard deviation of all entries.

```python
## Using df.values.max() will give the largest value in the dataframe

df.values.max()
```

```python
## Using df.values.min() will give the  smallest value in the dataframe

df.values.min()
```

```python
## Using df.values.mean() will give the mean in the dataframe

df.values.mean()
```

```python
## Using df.values.std() will give the standard deviation in the dataframe

df.values.std()
```
